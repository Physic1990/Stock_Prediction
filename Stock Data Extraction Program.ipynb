```python
import pandas as pd
import yfinance as yf
from google.colab import files
# Define the stock symbol and timeframe
stock_symbol = "AAPL"
start_date = "2020-01-01"
end_date = "2022-01-01"

# Download historical stock prices using yfinance
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Reset the index to have the date as a separate column
stock_data.reset_index(inplace=True)
# Extract only the date part from the datetime index
stock_data['Date'] = stock_data['Date'].dt.date
# Save the data to an Excel file
excel_file_path = "stock_data.xlsx"
stock_data.to_excel(excel_file_path, index=False)

print(f"Stock data saved to {excel_file_path}")
```
